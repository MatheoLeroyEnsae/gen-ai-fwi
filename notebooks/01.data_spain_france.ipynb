```python
# France 
# min_lon = -5
# max_lon = 9
# min_lat = 42
# max_lat = 51

# Spain 
# min_lon = -10
# max_lon = 2
# min_lat = 36
# max_lat = 44

MIN_LON = -5
MAX_LON = 9
MIN_LAT = 42
MAX_LAT = 51

S3_PATH = "s3://matheo/diffusion/data/data_all_years_france.parquet"
```

```python
import s3fs
import pandas as pd

MY_BUCKET = "matheo"
CHEMIN_FICHIER = "diffusion/data/data_1970_1983.parquet"
CHEMIN_FICHIER2 = "diffusion/data/data_1995_2005.parquet"
CHEMIN_FICHIER3 = "diffusion/data/data_1984_1994.parquet"
fs = s3fs.S3FileSystem(client_kwargs={"endpoint_url": "https://minio.lab.sspcloud.fr"})
with fs.open(f"s3://{MY_BUCKET}/{CHEMIN_FICHIER}") as f:
    data_1970_1983 = pd.read_parquet(f)
    data_1970_1983_filtered = data_1970_1983[(data_1970_1983["lon"]>MIN_LON) & (data_1970_1983["lon"]<MAX_LON) & (data_1970_1983["lat"]>MIN_LAT) & (data_1970_1983["lat"]<MAX_LAT)]
with fs.open(f"s3://{MY_BUCKET}/{CHEMIN_FICHIER2}") as f:
    data_1995_2005 = pd.read_parquet(f)
    data_1995_2005_filtered = data_1995_2005[(data_1995_2005["lon"]>MIN_LON) & (data_1995_2005["lon"]<MAX_LON) & (data_1995_2005["lat"]>MIN_LAT) & (data_1995_2005["lat"]<MAX_LAT)]

with fs.open(f"s3://{MY_BUCKET}/{CHEMIN_FICHIER3}") as f:
    data_1984_1994 = pd.read_parquet(f)
    data_1984_1994_filtered = data_1984_1994[(data_1984_1994["lon"]>MIN_LON) & (data_1984_1994["lon"]<MAX_LON) & (data_1984_1994["lat"]>MIN_LAT) & (data_1984_1994["lat"]<MAX_LAT)]
```

```python
final_df = pd.concat([data_1970_1983_filtered,data_1995_2005_filtered,data_1984_1994_filtered], ignore_index=True)
final_df.to_parquet(
    S3_PATH,
    engine="pyarrow",
)
```
